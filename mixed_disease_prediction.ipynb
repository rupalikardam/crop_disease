```python
import zipfile

zip_path = "/content/drive/MyDrive/Mixed Class Dataset.zip"
extract_path = "/content/"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset Extracted")
```

```python
import yaml

yaml_path = "/content/data.yaml"

with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

data['train'] = "/content/train/images"
data['val'] = "/content/valid/images"
data['test'] = "/content/test/images"

with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("YAML Updated")
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="/content/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    name="plant_disease_mixed"
)
```

```python
from IPython.display import display, HTML
import os
import base64

folder = "/content/runs/detect/predict"
images = os.listdir(folder)[:12]

html = ""

for img in images:
    path = os.path.join(folder, img)
    with open(path, "rb") as f:
        data = base64.b64encode(f.read()).decode()
    html += f'<img src="data:image/jpeg;base64,{data}" width="300" style="margin:5px;">'

display(HTML(html))
```
