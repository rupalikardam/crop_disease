**Class Count**

```python
import os
from collections import defaultdict

labels_path = "/content/Multi-Crop Disease Dataset/Multicrop Disease Dataset/Multicrop Disease Dataset/train/labels"   # change path if needed

class_counts = defaultdict(int)

for file in os.listdir(labels_path):
    with open(os.path.join(labels_path, file), "r") as f:
        lines = f.readlines()
        for line in lines:
            class_id = int(line.split()[0])
            class_counts[class_id] += 1

# Sort classes by frequency
sorted_classes = sorted(class_counts.items(), key=lambda x: x[1])

print("Class Distribution (Low → High):")
for cls, count in sorted_classes:
    print(f"Class {cls}: {count}")
```

```python
import os
import shutil

base_path = "/content/Multi-Crop Disease Dataset/Multicrop Disease Dataset/Multicrop Disease Dataset"

minority_path = "/content/minority_dataset"

folders = [
    "train/images", "train/labels",
    "valid/images", "valid/labels",
    "test/images", "test/labels"
]

for folder in folders:
    os.makedirs(os.path.join(minority_path, folder), exist_ok=True)
```

```python
minority_classes = [10, 11, 18, 12, 17]

splits = ["train", "valid", "test"]

for split in splits:
    label_dir = f"/content/Multi-Crop Disease Dataset/Multicrop Disease Dataset/Multicrop Disease Dataset/{split}/labels"
    image_dir = f"/content/Multi-Crop Disease Dataset/Multicrop Disease Dataset/Multicrop Disease Dataset/{split}/images"

    for file in os.listdir(label_dir):
        label_path = os.path.join(label_dir, file)

        with open(label_path, "r") as f:
            lines = f.readlines()

        keep = False

        for line in lines:
            class_id = int(line.split()[0])
            if class_id in minority_classes:
                keep = True
                break

        if keep:
            # copy label
            shutil.copy(label_path, f"/content/minority_dataset/{split}/labels/")

            # copy image
            img_name = file.replace(".txt", ".jpg")
            shutil.copy(
                os.path.join(image_dir, img_name),
                f"/content/minority_dataset/{split}/images/"
            )
```

```python
class_mapping = {
    10:0,
    11:1,
    18:2,
    12:3,
    17:4
}
```

```python
import os

label_dirs = [
    "/content/minority_dataset/train/labels",
    "/content/minority_dataset/valid/labels",
    "/content/minority_dataset/test/labels"
]

for label_dir in label_dirs:
    for file in os.listdir(label_dir):
        file_path = os.path.join(label_dir, file)

        new_lines = []

        with open(file_path, "r") as f:
            lines = f.readlines()

        for line in lines:
            parts = line.split()
            class_id = int(parts[0])

            if class_id in class_mapping:
                parts[0] = str(class_mapping[class_id])
                new_lines.append(" ".join(parts) + "\n")

        with open(file_path, "w") as f:
            f.writelines(new_lines)
```
